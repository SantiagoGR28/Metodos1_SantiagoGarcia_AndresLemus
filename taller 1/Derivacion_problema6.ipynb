```python
import numpy as np
import matplotlib.pyplot as plt
"""ejes"""
ejex=np.linspace(-4,4,25)
ejey=np.linspace(-4,4,25)
"""definir funcion"""
def potencial_flujo(x,y):
    vel=2
    radio=2
    flujo= (vel*x) * (1-((radio**2)/((x**2)+(y**2))))
    return flujo
"""derivadas"""
def DerX(f,x,y,h):  
    d = 0.
    if h != 0:
        d = (f(x+h,y) - f(x-h,y))/(2*h)
    return d
def DerY(f,x,y,h):
    d = 0.
    if h != 0:
        d = (f(x,y+h) - f(x,y-h))/(2*h)
    return d
"""check"""
Vx = np.zeros((25,25))
Vy = np.zeros((25,25))
h=0.001
for i in range(0,25):
    for j in range(0,25):
        if (ejex[i]**2 + ejey[j]**2) > 4:
            Vx[i,j] = DerX(potencial_flujo,ejex[i],ejey[j],h)
            Vy[i,j] = DerY(potencial_flujo,ejex[i],ejey[j],h)
        else:
            Vx[i,j] = 0
            Vy[i,j] = 0
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.plot(2*np.cos(np.linspace(0,2*np.pi)), 2*np.sin(np.linspace(0,2*np.pi)), color='r')
for i in range(0,25):
    for j in range(0,25):
        ax.quiver(ejex[i], ejey[j], Vx[i,j], Vy[i,j], color='b')
plt.show()
```
